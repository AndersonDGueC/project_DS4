```python
dic_ord_pr={}
column_new_ids=[]
keys_id=list(new_df_olist_order['order_id'].unique())




new_id=df_olist_orders[df_olist_orders['order_id'].isin(keys_id)]
list_ids=list(new_df_olist_order['order_id'])
#l1=len(new_df_olist_order['order_id'])
#l2=len(new_id['order_id'])
#print(l1)
#print(len(list_ids))
for count2, ids_ in enumerate(list_ids):
    dic_ord_pr[ids_]=new_df_olist_order['product_category_name'][count2]
    
for  keys1 in list(new_id['order_id']):
    column_new_ids.append(dic_ord_pr[keys1])

new_df_id=new_id.copy()
new_df_id['product_category_name']=column_new_ids

new_df_id["order_delivered_carrier_date"]=pd.to_datetime(new_df_id['order_delivered_carrier_date'])
new_df_id['order_delivered_customer_date']=pd.to_datetime(new_df_id['order_delivered_customer_date'])
new_df_id['order_estimated_delivery_date']=pd.to_datetime(new_df_id['order_estimated_delivery_date'])
new_df_id['order_approved_at']=pd.to_datetime(new_df_id['order_approved_at'])
new_df_id['order_purchase_timestamp']=pd.to_datetime(new_df_id['order_purchase_timestamp'])
new_df_id['difference']=new_df_id['order_delivered_customer_date']-new_df_id['order_delivered_carrier_date']
#len(dic_ord_pr.keys())
#len(list_ids)
new_df_id
```

```python
new_df_id['difference']=new_df_id['difference'].dt.days
```

```python
mean_day=prd_mean=new_df_id.groupby(['product_category_name'])['difference'].mean().reset_index(name="mean_days").sort_values('mean_days', ascending=True)
mean_day
```

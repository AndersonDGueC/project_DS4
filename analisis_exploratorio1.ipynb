```python
#DataSet Retailer Pricing
# 15- UK_retail_transactions.csv
df_uk_retailer = pd.read_csv("./e-commerce/datasets/UK_retail_transactions.csv")
df_uk_retailer["invoice_date"] = pd.to_datetime(df_uk_retailer["invoice_date"])
df_uk_retailer["Year"] = df_uk_retailer["invoice_date"].dt.year
df_uk_retailer["Month"] = df_uk_retailer["invoice_date"].dt.month
df_uk_retailer["Total_Value"] = df_uk_retailer["price"] * df_uk_retailer["quantity"]
df_uk_retailer["Order_Status"]= np.where((df_uk_retailer["quantity"] <= 0)|(df_uk_retailer["price"] <= 0), "Returned", "Delivered")
df_uk_retailer.head()
```

```python
df_uk_retailer.describe().round(2)
```

```python
#Remember to get out of reimburses (negative quantities)
plt.figure(figsize=(16, 6))
plt.ticklabel_format(style="plain", axis="y")
df_uk_retailer[(df_uk_retailer["price"] >= 0) | (df_uk_retailer["quantity"] >= 0)]["price"].hist();
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(x="Month", y="quantity", data=df_uk_retailer[(df_uk_retailer["quantity"] >= 0) | (df_uk_retailer["price"] >= 0)]);
```

```python
top_ten_uk_retailer = df_uk_retailer[(df_uk_retailer["quantity"] >= 0) | (df_uk_retailer["price"] >= 0)].groupby("country")["quantity"].size().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(16, 6))
plt.ticklabel_format(style="plain", axis="x")
sns.barplot(x="quantity", y="country", data=top_ten_uk_retailer, orient="h", ci=None);
```

```python
top_ten_uk_retailer_product = df_uk_retailer[(df_uk_retailer["quantity"] >= 0) | (df_uk_retailer["price"] >= 0)].groupby("description")["quantity"].size().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(16, 6))
sns.barplot(x="quantity", y="description", data=top_ten_uk_retailer_product, orient="h", ci=None);
```

```python
import warnings
warnings.filterwarnings("ignore")
categories = list(top_ten_uk_retailer_product["description"])
countries = ["United Kingdom", "EIRE", "Germany", "France", "Spain"]
top_ten_uk_value = df_uk_retailer[(df_uk_retailer["Order_Status"]=="Delivered") & (df_uk_retailer["description"].isin(categories))].groupby(["country","description"])["Total_Value"].size().sort_values(ascending=False).reset_index()
graph = sns.FacetGrid(top_ten_uk_value[top_ten_uk_value["country"].isin(countries)], col="country", palette="Set3", aspect=0.5, size=9)
graph.map(sns.barplot,"description","Total_Value")
graph.set_xticklabels(rotation=90)
plt.show();
```
